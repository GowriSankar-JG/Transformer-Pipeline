# Import Transformer Pipeline

```
from transformers import pipeline

text_classif_ppl = pipeline('sentiment-analysis')
qa_ppl = pipeline('question-answering')
summarize_ppl = pipeline('summarization')
text_gen_ppl = pipeline('fill-mask')
```

# 1. Sentiment Analysis

```
text = "the movie was incredible"
text_classif_ppl(text)
```

```
text = "the support team is useless"
text_classif_ppl(text)
```

<br>

#2. Question Answering

```
text = "The Mariana Trench is located in the East of the Mariana islands in the Western pacific Ocean"
ques = "Where is Mariana Trench?"

qa_ppl(context = text, question = ques)
```

<br>

#3. Text Summarization

```
text = """
        A yawl is similar to a ketch; whereas the mizzen mast is forward 
        of the rudder stock in the case of a ketch, it is aft of this point in the yawl rig. 
        Additionally, the mizzen sail tends to be significantly smaller relative to the 
        mainsail for the yawl compared to the ketch, about one quarter the size of the mainsail, 
        compared to the mizzen sail of a ketch, which may be about half the size of the mainsail.
        """

summary = summarize_ppl(text)

print("Summary:", summary[0]['summary_text'])
print("length of text:",len(text))
print("length of summary:",len(summary[0]['summary_text']))
```

<br>

#4. Text Generation

```
text = "I like to "

gen_text = text_gen_ppl(text + text_gen_ppl.tokenizer.mask_token) 

gen_text
```
